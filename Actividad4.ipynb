**Actividad 4 (Evaluación) (Valores Atípicos)**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Gastos_y_costos_20_23.xlsx')
data.head()
```

**Paso 1 - Verificación de datos nulos.**  
En este proceso obtenemos una visualización sobre la ausencia de datos en las variables del data frame.

```python
valores_nulos=data.isnull().sum()
valores_nulos
```

Dado que separaremos el data frame en 2 (datos categóricos y numéricos) se optó por utilizar el mismo tratamiento en datos nulos el cual será bfill.

```python
data1 =data.fillna(method="bfill") 
data1
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

Nos perctamaos que los campos se completan de manera adecuada

**Paso 2 - Identificación de outliers .**

Nos percatamos que en total contamos con un total de 8 variables numéricas, por lo cual pasaremos a realizar una extracción de estas con la finalidad de aplicar un filtro de detección de outliers.

```python
data_num=data1.iloc[ : , [8,9,10,11,12,13,14,18]]  #Columnas no consecutivas
data_num.info()
```

**Procedimiento de detección de outliers mediante el método de cuartiles**

Establecemos los rangos intercuartílicos.

```python
y=data_num

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

Vizualisamos los valores atípicos.

```python
data3_iqr= data_num[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

Procedemos a realizar una copia del dataframe con los intervalos cuartílicos y sustituimos los Na con el promedio de la variable analizada.

```python
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

Verificamos que nuestro DF no cuente con datos nulos.

```python
valores_nulos=data4_iqr.isnull().sum()
valores_nulos
```

Finalmente unimos los DF de datos categóricos y numéricos.

```python
df_limpio = pd.concat([data4_iqr, data1], axis=1)
df_limpio.isnull().sum()
```

Guardamos los valores limpios para un posterior análisis.

```python
df_limpio.to_csv("Datos_limpios.csv")
```
